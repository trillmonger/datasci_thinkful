```python
import math
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Load the dataset
PATH = 'https://raw.githubusercontent.com/trillmonger/datasci_thinkful/main/WA_Fn-UseC_-HR-Employee-Attrition.csv'
attrition_IBM = pd.read_csv(PATH)

# Preview the Data
attrition_IBM.info()
attrition_IBM.head(30)

#Income Variables & Exploration

High_Income = attrition_IBM[attrition_IBM['MonthlyIncome'] >= 7500]
Low_Income = attrition_IBM[attrition_IBM['MonthlyIncome'] <= 7500]


print(stats.describe(High_Income['MonthlyIncome']))

print(stats.describe(Low_Income['MonthlyIncome']))



#Histograms Checking for normality for Monthly Income 

plt.figure(figsize=(12, 6))

#High Income Employee Histogram
plt.subplot(1,2,1)
plt.hist(High_Income['MonthlyIncome'], bins=10, color='green', alpha=.5)
plt.title('Monthly Rate of High Income Employees')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')

#Low Income Employee Histogram
plt.subplot(1,2,2)
plt.hist(Low_Income['MonthlyIncome'], bins=10, color='red', alpha=.5)
plt.title('Monthly Rate of Low Income Employees')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
```

```python
#Job Satisfaction Variables & Exploration

High_Satisfy = attrition_IBM[attrition_IBM['JobSatisfaction'] >= 3]
Low_Satisfy = attrition_IBM[attrition_IBM['JobSatisfaction'] <= 2]


print(stats.describe(High_Satisfy['JobSatisfation']))

print(stats.describe(Low_Satisfy['JobSatisfation']))


#Histograms Checking for normality

plt.figure(figsize=(12, 6))

#Employees with High Job Statisfaction
plt.subplot(1,2,1)
plt.hist(High_Satisfy['JobSatisfaction'], bins=5, color='green', alpha=.5)
plt.title('Employees With High Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Frequency')

#Employees with Low Job Statisfaction
plt.subplot(1,2,2)
plt.hist(Low_Satisfy['JobSatisfaction'], bins=5, color='red', alpha=.5)
plt.title('Employees With Low Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Frequency')
```

```python
#Create Attrtition Variables for T-test and find P-value

#Income & Histogram for normality 

HighMula_stay = High_Income[High_Income['Attrition'] == 'No']

HighMula_leave = High_Income[High_Income['Attrition'] == 'Yes']

LowMula_stay = Low_Income[Low_Income['Attrition'] == 'No']

LowMula_leave = Low_Income[Low_Income['Attrition'] == 'Yes']

#Histogram for Attrition - High Income

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.hist(HighMula_stay['MonthlyIncome'], bins=10, color='green', alpha=.5, label = 'Attrition No')
plt.legend(loc="upper right")
plt.title('Attrition Rate of High Income Employees')
plt.xlabel('Monthly Income')
plt.ylabel('# of Employees')

plt.hist(HighMula_leave['MonthlyIncome'], bins=10, color='orange', alpha=.5, label = 'Attrition Yes')
plt.legend(loc="upper right")
plt.title('Attrition Rate of High Income Employees')
plt.xlabel('Monthly Income')
plt.ylabel('# of Employees')

#Histogram for Attrition - Low Income

plt.subplot(1,2,2)
plt.hist(LowMula_stay['MonthlyIncome'], bins=10, color='green', alpha=.5, label = 'Attrition No' )
plt.legend(loc="upper right")
plt.title('Attrition Rate of Low Income Employees')
plt.xlabel('Monthly Income')
plt.ylabel('# of Employees')

plt.hist(LowMula_leave['MonthlyIncome'], bins=10, color='orange', alpha=.5, label = 'Attrition Yes')
plt.legend(loc="upper right")
plt.title('Attrition Rate of Low Income Employees')
plt.xlabel('Monthly Income')
plt.ylabel('# of Employees')

#Job Satisfaction & Histogram for Normality

highJob_stay = High_Satisfy[High_Satisfy['Attrition'] == 'No']

highJob_leave = High_Satisfy[High_Satisfy['Attrition'] == 'Yes']

lowJob_stay = Low_Satisfy[Low_Satisfy['Attrition'] == 'No']

lowJob_leave = Low_Satisfy[Low_Satisfy['Attrition'] == 'Yes']


#Histogram for Attrition - High Job Satisfaction 

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.hist(highJob_stay['JobSatisfaction'], bins=10, color='green', alpha=.5, label = 'Attrition No')
plt.legend(loc="upper right")
plt.title('Attrition Rate for Satisfied Employees')
plt.xlabel('Job Satisfaction')
plt.ylabel('# of Employees')

plt.hist(highJob_leave['JobSatisfaction'], bins=10, color='orange', alpha=.5, label = 'Attrition Yes')
plt.legend(loc="upper right")
plt.title('Attrition Rate for Satisfied Employees')
plt.xlabel('Job Satisfaction')
plt.ylabel('# of Employees')


#Histogram for Attrition - Low Job Satisfaction 


plt.subplot(1,2,2)
plt.hist(lowJob_stay['JobSatisfaction'], bins=10, color='green', alpha=.5, label = 'Attrition No')
plt.legend(loc="upper right")
plt.title('Attrition Rate for Disatisfied Employees')
plt.xlabel('Job Satisfaction')
plt.ylabel('# of Employees')

plt.hist(lowJob_leave['JobSatisfaction'], bins=10, color='orange', alpha=.5, label = 'Attrition Yes')
plt.title('Attrition Rate for Disatisfied Employees')
plt.legend(loc="upper right")
plt.xlabel('Job Satisfaction')
plt.ylabel('# of Employees')

#T-test for 'MonthlyIncome'

t_stat_HighIncome, p_val_HighIncome = stats.ttest_ind(HighMula_stay['MonthlyIncome'],HighMula_leave['MonthlyIncome'])
                                                     
print(f'T-test for High Income Employees: t-statistic = {t_stat_HighIncome}, p-value = {p_val_HighIncome}.')

t_stat_LowIncome, p_val_LowIncome = stats.ttest_ind(LowMula_stay['MonthlyIncome'],LowMula_leave['MonthlyIncome'])

print(f'T-test for Low Income Employees: t-statistic = {t_stat_LowIncome}, p-value = {p_val_LowIncome}.')

#T-test for Job Satisfaction 

t_stat_HighJob , p_val_HighJob = stats.ttest_ind(highJob_stay['JobSatisfaction'],highJob_leave ['JobSatisfaction'])

print(f'T-test for Employees with high job satisfaction: t-statistic = {t_stat_HighJob}, p-value = {p_val_HighJob}.')

t_stat_LowJob , p_val_LowJob = stats.ttest_ind(lowJob_stay['JobSatisfaction'],lowJob_leave['JobSatisfaction'])

print(f'T-test for Employees with low job satisfaction: t-statistic = {t_stat_LowJob}, p-value = {p_val_LowJob}.')
```
